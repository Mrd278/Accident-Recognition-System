```
from keras.models import load_model
model = load_model("Accident_Recognition.h5")
model.summary()
```

```
from keras.preprocessing import image
X_train = []
img = image.load_img('1/'+'1.jpg',target_size = (32,32,3))
```

```
img = image.img_to_array(img)
img = img/255
X_train.append(img)
```

```
import numpy as np
X_train = np.asarray(X_train)
pred = model.predict(X_train)
```

```
print(pred)
```

```

np.round(pred) == 0
```
