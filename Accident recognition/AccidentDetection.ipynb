```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from keras.layers import Dense, Dropout, Flatten
from keras.models import Sequential
from keras.layers import MaxPooling2D, Conv2D
from keras.utils import np_utils
from tqdm import tqdm
from keras.preprocessing import image
```

```
model = Sequential()

model.add(Conv2D(filters = 16, kernel_size = (5,5), activation = 'relu',input_shape = (32,32,3)))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))

model.add(Conv2D(filters = 32, kernel_size = (5,5), activation = 'relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))

model.add(Flatten())

model.add(Dense(128,activation = 'relu'))
model.add(Dropout(0.25))

model.add(Dense(64,activation = 'relu'))
model.add(Dropout(0.25))

model.add(Dense(32,activation = 'relu'))
model.add(Dropout(0.25))

model.add(Dense(1,activation = 'sigmoid'))
```

```
model.summary()
```

```
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```
train_image = []
for i in tqdm(range(2500)):
  img = image.load_img('1/'+str(i+1)+'.jpg',target_size = (32,32,3))
  img = image.img_to_array(img)
  img = img/255
  train_image.append(img)
```

```
for i in tqdm(range(2979,4206)):
  img = image.load_img('2/'+str(i+1)+'.jpg',target_size = (32,32,3))
  img = image.img_to_array(img)
  img = img/255
  train_image.append(img)
```

```
X_train = np.array(train_image)
X_train.shape
```

```
dataset = X_train
y = np.zeros(4896)

y[2501:4896] = np.ones(2395)
```

```
y.shape
```

```
X_train, X_test, y_train, y_test = train_test_split(dataset,y,test_size = 0.15,random_state = 0)
```

```
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```
model.fit(X_train,y_train, epochs = 50, validation_data = (X_test,y_test), batch_size = 16)
```

```
model_name = "Accident_Recognition.h5"
model.save(model_name,overwrite = True)
```
